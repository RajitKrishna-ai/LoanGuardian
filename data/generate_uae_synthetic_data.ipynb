```python
import pandas as pd
import numpy as np

np.random.seed(42)

n_samples = 50000

# UAE-style synthetic features
df = pd.DataFrame({
    'LoanID': range(1, n_samples+1),
    'LoanAmount': np.random.randint(5000, 1000000, n_samples),  # AED
    'Income': np.random.randint(2000, 40000, n_samples),
    'Age': np.random.randint(21, 65, n_samples),
    'CreditScore': np.random.randint(300, 850, n_samples),
    'ExistingEMI': np.random.randint(0, 20000, n_samples),
    'EmploymentStatus': np.random.choice(['Employed', 'Self-Employed', 'Unemployed', 'Retired'], n_samples),
    'Tenure': np.random.randint(6, 60, n_samples),
    'MaritalStatus': np.random.choice(['Single', 'Married', 'Divorced', 'Widowed'], n_samples),
    'Dependents': np.random.randint(0, 5, n_samples),
    'RepaymentHistory': np.random.choice(['Good', 'Average', 'Poor'], n_samples, p=[0.7,0.2,0.1]),
    'Purpose': np.random.choice(['Home Loan', 'Car Loan', 'Personal Loan', 'Business Loan'], n_samples),
    'Stage': np.random.choice(['Lead', 'File', 'Sanction', 'Disbursement'], n_samples),
    'DelinquencyCount12M': np.random.randint(0, 5, n_samples)
})

# Generate synthetic target
df['DefaultStatus'] = ((df['CreditScore'] < 550) & (df['RepaymentHistory'] == 'Poor')).astype(int)

# Save to CSV
df.to_csv('loan_guardian_uae.csv', index=False)
print("Synthetic UAE dataset created: loan_guardian_uae.csv")
```
